# Зависимости + считывание данных

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

data_1 = np.loadtxt('VD5.txt')
data_2 = np.loadtxt('VD6.txt')
data_3 = np.loadtxt('VD7.txt')
data_4 = np.loadtxt('HL1.txt')
```

# ВАХ VD5

```python
U, I = data_1[:, 0], data_1[:, 1]
plt.figure(figsize=(10, 6))
plt.grid()
plt.xlabel('U_VD5, В')
plt.ylabel('I_VD5, мА')
plt.plot(U, I)
plt.savefig('graphs/ВАХ_VD5.png', dpi=300)
```

# ВАХ VD6

```python
U, I = data_2[:, 0], data_2[:, 1]
plt.figure(figsize=(10, 6))
plt.grid()
plt.xlabel('U_VD6, В')
plt.ylabel('I_VD6, мА')
plt.plot(U, I)
plt.savefig('graphs/ВАХ_VD6.png', dpi=300)
```

# ВАХ VD7

```python
U, I = data_3[:, 0], data_3[:, 1]
dU, dI = np.diff(U), np.diff(I)
R_dif = dU / dI * 1000
plt.figure(figsize=(10, 6))
plt.grid()
plt.xlabel('U_VD7, В')
plt.ylabel('I_VD7, мА')
plt.plot(U, I)
plt.savefig('graphs/ВАХ_VD7.png', dpi=300)
data = {"dU, В": dU, "dI, мА": dI, "R_dif, Ом": R_dif}
df = pd.DataFrame(data)
df = df.round(3)
df
```

# ВАХ HL1

```python
U, I, U_1 = data_4[:, 0], data_4[:, 1], data_4[:, 2]
U_HL1 = U - (I * 500 / 1000) # RB2 = 500ом
U_1 = U_1 - (I * 500 / 1000)
plt.figure(figsize=(10, 6))
plt.grid()
plt.xlabel('U, В')
plt.ylabel('I_HL1, мА')
plt.plot(U_HL1, I, label='1. I_HL1=f(U_HL1), U=U_HL1')
plt.plot(U_1, I, label='2. I_HL1=f(U_RB2+HL1), U=U_RB2+HL1')
plt.legend()
plt.savefig('graphs/ВАХ_HL1.png', dpi=300)
```
